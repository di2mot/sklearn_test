The **KMeans algorithm** employs clustering techniques to group data by attempting to segregate samples into 'n' clusters with equal variance. The optimization criterion it minimizes is known as *inertia* or *within-cluster sum-of-squares* (refer to the [documentation](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans) for details). To use this algorithm effectively, it is essential to specify the number of clusters beforehand. Notably, KMeans demonstrates scalability to handle extensive datasets and has found application in diverse fields, showcasing its versatility and effectiveness.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

First, let's generate an array of 100 random points in two-dimensional space divided into two groups.


```python
X = -2 * np.random.rand(100, 2)
X1 = 1 + 2 * np.random.rand(50, 2)
X[50:100, :] = X1
plt.scatter(X[:, 0], X[:, 1], s=50, c= 'b')
plt.show()
```

Since we have two groups of points, the number of clusters is set to 2 (n_clusters = 2).

```python
kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(X)
kmeans.labels_
```

Coordinates of the found centers

```python
kmeans.cluster_centers_
```

Let's overlay them with clusters

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50, c='b')
plt.scatter(-0.94665068, -0.97138368, s=200, c='g', marker='s')
plt.scatter(2.01559419, 2.02597093, s=200, c='r', marker='s')
plt.show()
```

Let's make sure that the algorithm works correctly and our points are distributed over two points.

```python
kmeans.labels_
```

Let's test the algorithm. Give it the coordinate of a point to determine whether it belongs to a cluster.

```python
sample_test=np.array([-3.0,-3.0])
second_test=sample_test.reshape(1, -1)
kmeans.predict(second_test)
```
